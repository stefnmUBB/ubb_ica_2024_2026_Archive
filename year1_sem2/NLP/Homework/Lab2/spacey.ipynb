```python
sentences_en = [
    'She liked to pickle all her vegetables, even carrots and onions.',
    'When you finish your rice, you can eat dessert.',
    'Who wrote “Treasure Island”?',
    'If clouds had souls it would never stop raining.',
    'I watched the people in the park while I sat on the swing, and now I want to write them into a story.',
    'This is going to be the stupidest birthday I\'ve ever had.',
    'The llama couldn\'t resist trying the lemonade.',
    'I purchased a baby clown from the Russian terrorist black market.',
    'He said he was not there yesterday; however, many people saw him there.',
    'As the asteroid hurtled toward earth, Becky was upset her dentist appointment had been canceled.'
]
```

```python
import spacy
from spacy import displacy
from tabulate import tabulate
```

```python
nlp_ro = spacy.load("ro_core_news_sm")
nlp_en = spacy.load('en_core_web_sm')
```

```python
def process(nlp, sentence, i):
    print("_______________________________________________________")
    print(f"{i}. {sentence}")
    print("_______________________________________________________")
    doc = nlp(sentence)
    table = [['text', 'lemma', 'pos', 'tag', 'dep', 'shape', 'is_alpha', 'is_stop']] 
    for token in doc:
        table.append([token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop])
    print(tabulate(table, headers='firstrow'))

    table_ents = [['entity', 'label']]
    for ent in doc.ents:
        table_ents.append([str(ent), ent.label_])
    print("NER:")
    print(tabulate(table_ents, headers='firstrow'))

    displacy.render(doc, style="dep")
    
```

```python
for i in range(len(sentences_ro)):
    process(nlp_ro, sentences_ro[i], i+1)
```

```python
for i in range(len(sentences_en)):
    process(nlp_en, sentences_en[i], i+1)
```
